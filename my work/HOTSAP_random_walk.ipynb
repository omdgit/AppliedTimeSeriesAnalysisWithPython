# Random Walk 

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

import sys
sys.path.append('../../My-Code/')

import timetools as tt

%matplotlib inline
```

```python
# plt.rcParams['figure.figsize'] = [10, 7.5]
```

```python
np.random.seed(1234)

steps = np.random.standard_normal(1000)
steps[0]=0
random_walk = np.cumsum(steps)
```

```python
random_walk[:10]
```

```python
tt.get_plot_styles(random_walk)
```

```python
plt.style.use('seaborn')
plt.plot(random_walk)
plt.title("Simulated Random Walk")
plt.show()
```

```python
tt.plot_acf_pcf(random_walk, num_lag=20)
```

```python
random_walk_acf_coef = acf(random_walk)
random_walk_acf_coef
```

```python
plot_acf(random_walk, lags=20);
```

```python

random_walk_diff = np.diff(random_walk, n=1)
df_rw = pd.DataFrame(random_walk, columns=['rw'])
```

```python
# plt.figure(figsize=[10, 7.5]); # Set dimensions for figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 5))
ax1.plot(random_walk_diff)
ax1.set_title('Noise')
ax2.plot(df_rw.diff(1))
ax2.set_title('Noise')
plt.show()
```

```python
plot_acf(random_walk_diff, lags=20);
```

```python
sns.distplot(random_walk_diff, hist=True, kde=True, bins=20);
```
